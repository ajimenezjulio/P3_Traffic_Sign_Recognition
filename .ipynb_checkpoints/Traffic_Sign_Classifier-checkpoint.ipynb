### Include an exploratory visualization of the dataset

````python
def get_class_names(file_path):
    """
    Get class names of all the classes
    
    Parameters:
        - file_path: Path of the csv file
        
    Output:
        - array with the classnames
    """
    classnames = []
    with open(file_path) as _f:
        rows = csv.reader(_f, delimiter=',')
        # Skip headers
        next(rows, None)
        for i, row in enumerate(rows):
            assert(i == int(row[0]))
            # Append class name
            classnames.append(row[1])
    
    return classnames


def get_image_from_class(img_path, c): 
    """
    Returns the image matching the corresponging class
    
    Parameters:
        - img_path: Path of the image that has all the signs
        - c: Class number 
        
    Output:
        - image matching the given class
    """
    # Load the image in RGB
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # 7 images per row and 7 rows
    height, width, _ = img.shape
    dH = height / 7.
    dW = width / 7.104
    y = c // 7
    x = c % 7 
    
    # Extract image
    label_image = img[round(y * dH) : round(y * dH + dH), round(x * dW) : round(x * dW + dW), :]
    label_image = cv2.resize(label_image, (0,0), fx = 32. / dW, fy = 32. / dH)
    return label_image


def insert_subimage(image, sub_image, y, x): 
    """
    Inserts a subimage in a given image
    
    Parameters:
        - image: Parent image
        - sub_image: child image to be inserted
        - y: y position
        - x: x position
        
    Output:
        - image after insertion
    """
    height, width, c = sub_image.shape
    image[y : y + height, x : x + width, :] = sub_image 
    return image


def create_image_of_dataset(img_path, train_images, labels, class_names, n_samples):
    """
    Returns the summary image of the dataset
    
    Parameters:
        - img_path: Path of the image that has all the signs
        - train_images: Images in the training dataset
        - labels: labels of the train_images
        - class_names: Array with the names of each class
        - n_samples: Number of samples to insert per class
        
    Output:
        - dataset_image: Contains the original class image, average class image, 
                         sample images of that class and number of images
    """
    _, height, width, channel = train_images.shape
    n_classes = len(class_names)
    
    # Output image
    dataset_image = 255. * np.ones(shape =(n_classes * height,(n_samples + 24) * width, channel), dtype=np.float32)

    for c in range(n_classes):
        # Extract image of the class and insert it into dataset_image
        label_image = get_image_from_class(img_path, c)
        insert_subimage(dataset_image, label_image, c * height, 0)

        # Averaging image
        idx = list(np.where(labels == c)[0])
        mean_image = np.average(train_images[idx], axis = 0)
        insert_subimage(dataset_image, mean_image, c * height, width)

        # Random sampling
        for n in range(n_samples):
            sample_image = train_images[np.random.choice(idx)]
            insert_subimage(dataset_image, sample_image, c * height, (2 + n) * width)

        # Summary
        count = len(idx)
        color = [int(_) for _ in tuple(np.random.choice(range(255), size=3))]
        
        percentage = float(count) / float(len(train_images))
        cv2.putText(dataset_image, 
                    ' {:02d}: {:6}'.format(c, class_names[c]), 
                    ((2 + n_samples) * width, int((c + 0.7) * height)),
                    cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),1, 
                    cv2.LINE_AA)
        cv2.putText(dataset_image, 
                    '{:4d}'.format(count), 
                    ((2 + n_samples + 14) * width, int((c+0.7) * height)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
        cv2.rectangle(dataset_image,
                      ((2 + n_samples + 16) * width, c * height),
                      ((2 + n_samples + 16) * width + round(percentage * 3000), (c+1)*height),
                      color,
                      -1)
        
        #cv2.imwrite(DATA_DIR+'/data_summary.jpg',cv2.cvtColor(results_image, cv2.COLOR_BGR2RGB))
    
    return dataset_image
    
    
````

## Pre-process the Data Set

Every batch will be normalized before feeding the network using a tensorflow batch normalization function, in this section we will cover how we extend the dataset to generate more data for the model. The functions are the following ones:

1. Extend data by flipping 
1. Balance the data (all classes will have the same number of samples)
1. Perturb the image by warping and modifying brightness, contrast and saturation


````python
# Taken from http://navoshta.com/traffic-signs-classification/
def extend_data_by_flipping(images, labels, n_classes):
    """
    Extends the dataset by flipping it horizontally, vertically or both
    taking into account the relationships between the classes
    after the rotation.
    
    Parameters:
        - images: Images in the training dataset
        - labels: Labels of the training dataset
        - n_classes: Number of classes
        
    Output:
        - X_extended: Extended dataset of training images
        - y_extended: Extended dataset of training labels
    """
    
    X = images
    y = labels

    # Classes of signs that, when flipped horizontally, should still be classified as the same class
    self_flippable_horizontally = np.array([11, 12, 13, 15, 17, 18, 22, 26, 30, 35])
    # Classes of signs that, when flipped vertically, should still be classified as the same class
    self_flippable_vertically = np.array([1, 5, 12, 15, 17])
    # Classes of signs that, when flipped horizontally and then vertically, should still be classified as the same class
    self_flippable_both = np.array([32, 40])
    # Classes of signs that, when flipped horizontally, would still be meaningful, but should be classified as some other class
    cross_flippable = np.array([
        [19, 20],
        [33, 34],
        [36, 37],
        [38, 39],
        [20, 19],
        [34, 33],
        [37, 36],
        [39, 38],
    ])

    X_extended = np.empty([0, X.shape[1], X.shape[2], X.shape[3]], dtype = np.float32)
    y_extended = np.empty([0], dtype = np.int32)

    for c in range(n_classes):
        # First copy existing data for this class
        X_extended = np.append(X_extended, X[y == c], axis=0)
        # If we can flip images of this class horizontally and they would still belong to said class...
        if c in self_flippable_horizontally:
            # ...Copy their flipped versions into extended array.
            X_extended = np.append(X_extended, X[y == c][:, :, ::-1, :], axis=0)
        # If we can flip images of this class horizontally and they would belong to other class...
        if c in cross_flippable[:, 0]:
            # ...Copy flipped images of that other class to the extended array.
            flip_class = cross_flippable[cross_flippable[:, 0] == c][0][1]
            X_extended = np.append(X_extended, X[y == flip_class][:, :, ::-1, :], axis=0)
        # Fill labels for added images set to current class.
        y_extended = np.append(y_extended, np.full((X_extended.shape[0] - y_extended.shape[0]), c, dtype=np.int32))

        # If we can flip images of this class vertically and they would still belong to said class...
        if c in self_flippable_vertically:
            # ...Copy their flipped versions into extended array.
            X_extended = np.append(X_extended, X_extended[y_extended == c][:, ::-1, :, :], axis=0)
        # Fill labels for added images set to current class.
        y_extended = np.append(y_extended, np.full((X_extended.shape[0] - y_extended.shape[0]), c, dtype=np.int32))

        # If we can flip images of this class horizontally AND vertically and they would still belong to said class...
        if c in self_flippable_both:
            # ...Copy their flipped versions into extended array.
            X_extended = np.append(X_extended, X_extended[y_extended == c][:, ::-1, ::-1, :], axis=0)
        # Fill labels for added images set to current class.
        y_extended = np.append(y_extended, np.full((X_extended.shape[0] - y_extended.shape[0]), c, dtype=np.int32))

    return (X_extended, y_extended)


def perturb(image, keep, angle_limit=15, scale_limit=0.1, translate_limit=3, distort_limit=3, illumin_limit=0.7):
    """
    Perturb an image (warping, brightness, contrast and saturation)
    based on a given probability (keep) and uses random uniform
    distribution for perturbed values.
    
    Parameters:
        - image: Image to be perturbed
        - keep: Probability of the image to not be perturbed
        - angle_limit: Limit of the rotation angle for the image
        - scale_limit: Limit of the scale for the image
        - translate_limit: Limit of the translation for the image
        - distort_limit: Limit of the distortion noise to be added
        - illumin_limit: Limit of the illumination for the image
        
    Output:
        - image: Either perturbed or original image
                 depending on the keep probability
    """
    u = np.random.uniform()
    
    if u > keep :
        (width, height, channel) = image.shape
        center = np.array([width / 2., height / 2.])
        da = np.random.uniform(low=-1, high=1) * angle_limit/180. * math.pi
        scale = np.random.uniform(low=-1, high=1) * scale_limit + 1

        cc = scale * math.cos(da)
        ss = scale * math.sin(da)
        rotation    = np.array([[cc, ss], [-ss, cc]])
        translation = np.random.uniform(low=-1, high=1, size=(1,2)) * translate_limit
        distort     = np.random.standard_normal(size=(4,2)) * distort_limit
        
        # Warping points
        pts1 = np.array([[0., 0.], [0., width], [width, height], [width, 0.]])
        pts2 = np.matmul(pts1-center, rotation) + center  + translation

        # Add perspective noise
        pts2 = pts2 + distort
        
        # Distortion matrix and distorted image
        matrix  = cv2.getPerspectiveTransform(pts1.astype(np.float32), pts2.astype(np.float32)) 
        perturb = cv2.warpPerspective(image, matrix, (width, height), flags=cv2.INTER_LINEAR,
                                      borderMode=cv2.BORDER_REFLECT_101)  # BORDER_WRAP  #BORDER_REFLECT_101  #cv2.BORDER_CONSTANT  BORDER_REPLICATE

        # Brightness
        if 1:  
            alpha = 1.0 + illumin_limit * random.uniform(-1, 1)
            perturb *= alpha
            perturb = np.clip(perturb, 0., 255.)
            pass
        
        # Contrast
        if 1:  
            # RGB to gray (YCbCr) :  Y = 0.299R + 0.587G + 0.114B
            coef = np.array([[[0.299, 0.587, 0.114]]]) 

            alpha = 1.0 + illumin_limit * random.uniform(-1, 1)
            gray = perturb * coef
            gray = (3.0 * (1.0 - alpha) / gray.size) * np.sum(gray)
            perturb *= alpha
            perturb += gray
            perturb = np.clip(perturb,0.,255.)
            pass
        
        # Saturation
        if 1:  
            coef = np.array([[[0.299, 0.587, 0.114]]])

            alpha = 1.0 + illumin_limit*random.uniform(-1, 1)
            gray = perturb * coef
            gray = np.sum(gray, axis=2, keepdims=True)
            gray *= (1.0 - alpha)
            perturb *= alpha
            perturb += gray
            perturb = np.clip(perturb,0.,255.)
            pass

        return perturb

    else:
        return image
    
    
def make_perturb_images(images, keep):
    """
    Perturb the given images
    
    Parameters:
        - images: Images to be perturbed
        - keep: Probability of the image to not be perturbed
        
    Output:
        - Perturbed images
    """
    perturbed_images = np.zeros(images.shape)
    for n in range(len(images)):
        perturbed_images[n] = perturb(images[n],keep = keep)
        
    return perturbed_images


def balance_data(images, labels, n_classes, num_per_class=None):
    """
    Balance and shuffle the data (i.e. same 
    number of samples in each class).
    
    Parameters:
        - images: Images in the training dataset
        - labels: Labels in the training dataset
        - n_classes: Total number of classes
        - num_per_class: Desired number of images per class
        
    Output:
        - shuffle_images: Balanced and shuffled images
        - shuffle_labels: Balanced and shuffled labels
    """
    # Get maximum number of images in a class
    if num_per_class is None:
        max_count = 0
        for c in range(n_classes):
            idx = list(np.where(labels == c)[0])
            count = len(idx)
            max_count = max(count, max_count)
        num_per_class = max_count

    index = []
    for c in range(n_classes):
        idx = list(np.where(labels == c)[0])
        index = index + list(np.random.choice(idx, num_per_class))

    random.shuffle(index)
    shuffle_images  = images[index]
    shuffle_labels = labels[index]

    return shuffle_images, shuffle_labels
````

````python
# Train data filpping.
train_images, train_labels = extend_data_by_flipping(X_train, y_train, n_classes) 
num_train_flip = len(train_images)
print('Number of training images after flipping: {:d}'.format(num_train_flip))

# Keep 20 percent of the data
keep = 0.20
# Balance to 2000 images per class
num_per_class = 2000
# Total of images in train set
total_images = num_per_class * n_classes

# Balance the data
train_balanced_images, train_balanced_labels = balance_data(train_images, 
                                                            train_labels, 
                                                            n_classes, 
                                                            num_per_class = num_per_class)
# Perturb 80% of images, 100-20 = 80
train_balanced_images = make_perturb_images(train_balanced_images, keep = keep) 
print('Total of images after balancing: {:d}'.format(total_images))

# Display graph of data balanced
get_data_stats(train_balanced_labels, y_test)
````

### Model Architecture

````python
#Modified LeNet Architecture

# conv1->relu->pool->dropout->conv2->relu->pool->dropout->conv3->relu->pool->dropout->fc1->rel->dropout->fc2

conv1_W = None;
conv2_W = None;
conv3_W = None
fc1_W = None;
fc2_W = None;

def LeNet(x):   
    """
    Creates every layer of the architecture of the neural network,
    this project consists of 3 convolutional layers and 2 dense
    layers.
    
    Parameters:
        - images: Images in the training dataset
        
    Output:
        - Logits of the neural network
    """
    
    global conv1_W,conv2_W, conv3_W, fc1_W, fc2_W
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # SOLUTION: Normalize dataset before feeding the network
    input = tf.contrib.layers.batch_norm(x, decay = 0.9, epsilon = 1e-5, center = True, scale = True,
                              is_training = 1,reuse = None, updates_collections = None)
    
    
    # SOLUTION: Layer 1: Convolutional. Input = 32x32x3. Output = 32x32x16.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 3, 16), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(16))
    conv1   = tf.nn.conv2d(input, conv1_W, strides=[1, 1, 1, 1], padding='SAME') + conv1_b
    print ('conv1-->', conv1)
    tf.add_to_collection('vars', conv1_W)
    
    # SOLUTION: Activation.
    conv1 = tf.nn.relu(conv1)
    print ('conv1 activation-->', conv1)
    # SOLUTION: Pooling. Input = 32x32x16. Output = 16x16x16.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
    print ('conv1 pooling -->', conv1)
    
    # Dropout
    conv1 = tf.nn.dropout(conv1, keep_prob_conv1)
    print ('conv1 dropout -->', conv1)
    
    
    
    # SOLUTION: Layer 2: Convolutional. Input = 16x16x16 Output = 16x16x32?
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 16, 32), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(32))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='SAME') + conv2_b
    print ('conv2-->', conv2)
    tf.add_to_collection('vars', conv2_W)
    
    # SOLUTION: Activation.
    conv2 = tf.nn.relu(conv2)
    print ('conv2 activation-->', conv2)
    
    # SOLUTION: Pooling. Input = 16x16x32. Output = 8x8x32.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
    print ('conv2 pooling-->', conv2)
    
    # Dropout
    conv2 = tf.nn.dropout(conv2, keep_prob_conv2)
    print ('conv2 dropout-->', conv2)
    
    
    
    # SOLUTION: Layer 3: Convolutional. Input =8x8x32 Output = 8x8x64?
    conv3_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 32, 64), mean = mu, stddev = sigma))
    conv3_b = tf.Variable(tf.zeros(64))
    conv3   = tf.nn.conv2d(conv2, conv3_W, strides=[1, 1, 1, 1], padding='SAME') + conv3_b
    print ('conv3-->', conv3)
    tf.add_to_collection('vars', conv3_W)
    
    # SOLUTION: Activation.
    conv3 = tf.nn.relu(conv3)
    print ('conv3 activation-->', conv3)

    # SOLUTION: Pooling. Input = 8x8x64. Output = 4x4x64.
    conv3 = tf.nn.max_pool(conv3, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
    print ('conv3 pooling-->', conv3)
    
    # Dropout
    conv3 = tf.nn.dropout(conv3, keep_prob_conv3)
    print ('conv3 pooling-->', conv3)
    
    
    
    # SOLUTION: Flatten. Input = 4x4x64. Output = 1024.
    fc0   = flatten(conv3)
    print ('flattened layer shape-->', fc0)
    
    # SOLUTION: Layer 3: Fully Connected. Input = 1024. Output = 84.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(1024, 84), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(84))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    print ('fc1 shape-->', fc1)
    tf.add_to_collection('vars', fc1_W)
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)
    print ('fc1 activation-->', fc1)
    
    #dropout
    fc1 = tf.nn.dropout(fc1, keep_prob_fc1)
    print ('fc1 dropout-->', fc1)

    
    
    # SOLUTION: Layer 5: Fully Connected. Input = 84. Output = 43.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(84, 43), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc1, fc2_W) + fc2_b
    print ('logits-->', logits)
    tf.add_to_collection('vars', fc2_W)
    tf.add_to_collection('vars', logits)
    
    return logits
````

sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
from sklearn.utils import shuffle
import matplotlib.pyplot as plt

# Training images and labels placeholders
x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)

# Probabilities to keep neurons in each layer
keep_prob_conv1 = tf.placeholder(tf.float32)
keep_prob_conv2 = tf.placeholder(tf.float32)
keep_prob_conv3 = tf.placeholder(tf.float32)
keep_prob_fc1 = tf.placeholder(tf.float32)

# Learning rate
rate = 0.0005

print('Summary CNN model\n')
print('----------------------------------------------------------------------------------------------')

# Logits
logits = LeNet(x)

# Neural network operations
# Using L2 regularizers
regularizers = tf.nn.l2_loss(conv1_W) + tf.nn.l2_loss(conv2_W) + tf.nn.l2_loss(conv3_W) + tf.nn.l2_loss(fc1_W) + tf.nn.l2_loss(fc2_W)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy + 0.00005 * regularizers)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)

# Predictions
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()
````

````python
def evaluate(X_data, y_data, batch_size):
    """
    Model accuracy evaluation function.
    
    Parameters:
        - X_data: Images
        - y_data: Labels
        - batch_size: Number of images per batch
        
    Output:
        - accuracy of the model
    """
    num_training = len(X_data)
    total_accuracy = 0
    
    sess = tf.get_default_session()
    for offset in range(0, num_training, batch_size):
        batch_x, batch_y = X_data[offset:offset + batch_size], y_data[offset:offset + batch_size]
        # Evaluate the accuracy, remove dropouts, we want to use all neurons for evaluation
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, 
                                                           y: batch_y, 
                                                           keep_prob_conv1: 1.0,
                                                           keep_prob_conv2: 1.0, 
                                                           keep_prob_conv3: 1.0, 
                                                           keep_prob_fc1: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_training


def plot_graph(epochs, accuracy, rate, batch_size, arch_name):
    """
    Displays and saves the graph of the model, epochs vs accuracy
    
    Parameters:
        - epochs: Number of epochs
        - accuracy: Accuracy array
        - rate: Learning rate
        - batch_size: Number of images per batch
        - arch_name: Name of the architecture        
    """
    fig1 = plt.figure(figsize=(10,6))
    plt.plot(epochs, accuracy)
    plt.xlim(0, len(epochs))
    plt.ylim(0, 1)
    plt.xlabel('epochs')
    plt.ylabel('validation - accuracy')
    plt.title(arch_name + '--> Validation Accuracy - Epoch')
    fig1.savefig('results/' + arch_name + "__" + str(rate) + "_" + str(batch_size) + ".jpeg", bbox_inches='tight')
    
````

````python
# Control variables
EPOCHS = 200
BATCH_SIZE = 128
ARCH_NAME = '3conv-2fc 3-16-32-64-1024-84-43'

# Train the model
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    # Array to store the accuracy of the validation set
    # at each epoch
    validation_acc = []
    # Current batch accuracy
    batch_acc = None
    
    num_training = len(train_balanced_images)
    
    print("Training model with", num_training, "images\n\n")
    print(' epoch   iter    |  batch_acc   |  valid_acc    ')
    print('------------------------------------------------')
    
    # Variable to store the current iteration
    iteration = 0
    
    for epoch in range(EPOCHS):
        for offset in range(0, num_training, BATCH_SIZE):
            iteration += 1
            fraction_epoch = float(iteration * BATCH_SIZE) / float(num_training)
            
            # Creating mini-batch set
            end = offset + BATCH_SIZE
            batch_x, batch_y = train_balanced_images[offset:end], train_balanced_labels[offset:end]
            
            # Run the training phase in the mini-batch
            sess.run(training_operation, feed_dict={x: batch_x, 
                                                    y: batch_y, 
                                                    keep_prob_conv1: 0.9, 
                                                    keep_prob_conv2: 0.6, 
                                                    keep_prob_conv3: 0.6, 
                                                    keep_prob_fc1: 0.5})
            
            # Evaluate accuracy in the current batch
            batch_acc = evaluate(batch_x, batch_y, BATCH_SIZE)
            print('\r{:5.1f}   {:06d}   |  {:f}    |'.format(fraction_epoch, 
                                                                  iteration, 
                                                                  batch_acc), end = '', flush = True)
         
        # Evaluate accuracy in the validation set after the epoch finalization
        validation_acc.append(evaluate(X_valid, y_valid, BATCH_SIZE))
        print('\r{:5.1f}   {:06d}   |  {:f}    |  {:f}  '.format(fraction_epoch, 
                                                                 iteration, 
                                                                 batch_acc,
                                                                 validation_acc[epoch]), end = '\n', flush = True)
    # Save the model and display the graph    
    saver.save(sess, 'Model/{}'.format(ARCH_NAME))
    print("Model saved")
    plot_graph(list(np.arange(1,(EPOCHS+1))), validation_acc, rate, BATCH_SIZE, ARCH_NAME)
    
````
